```python
import os
import pandas as pd
import numpy as np
import torch
from torch.utils.data import DataLoader, TensorDataset
from sklearn.preprocessing import MinMaxScaler
import pickle

# Dizin tanımlamaları
temiz_veri_ana_klasor = r"C:\Users\feyza\Desktop\uav_project\data\processed\cleaned_train" # Kendi temiz klasör yolun kanka
modeller_dir = r"C:\Users\feyza\Desktop\uav_project\models"

# 1. Modülün (IMU) hedef kolonları
imu_features = ['gyro_rad[0]', 'gyro_rad[1]', 'gyro_rad[2]', 'accelerometer_m_s2[0]', 'accelerometer_m_s2[1]', 'accelerometer_m_s2[2]']

print("⚡ 1. MODÜL: IMU Autoencoder Veri Hazırlama Motoru Başlatıldı...\n")

all_imu_data = []

if not os.path.exists(temiz_veri_ana_klasor):
    print("❌ Hata: Temiz uçuşların durduğu ana klasör bulunamadı!")
else:
    # Kararlı test rehberindeki mantıkla alt klasörleri tarıyoruz
    for kok_dizin, alt_dizinler, dosyalar in os.walk(temiz_veri_ana_klasor):
        for dosya in dosyalar:
            # Sadece sensor_combined dosyalarını hedef alıyoruz, performansı koruyoruz!
            if "sensor_combined" in dosya and dosya.endswith(".csv"):
                tam_yol = os.path.join(kok_dizin, dosya)
                try:
                    df_sub = pd.read_csv(tam_yol)
                    
                    # Eksik veri kontrolü ve temizliği
                    df_sub = df_sub[imu_features].dropna()
                    
                    if not df_sub.empty:
                        all_imu_data.append(df_sub)
                except Exception:
                    continue

if len(all_imu_data) > 0:
    # Master IMU matrisini oluşturma
    df_imu_master = pd.concat(all_imu_data, ignore_index=True)
    print(f"📊 Toplam {df_imu_master.shape[0]} satır pürüzsüz IMU verisi birleştirildi.")
    
    # MinMax Normalizasyon
    print("⚖️ Veriler [0, 1] aralığına normalize ediliyor...")
    imu_scaler = MinMaxScaler()
    scaled_imu = imu_scaler.fit_transform(df_imu_master)
    
    # Arayüzün anlık normalizasyon yapabilmesi için scaler'ı kaydediyoruz
    os.makedirs(modeller_dir, exist_ok=True)
    with open(os.path.join(modeller_dir, "imu_scaler.pkl"), "wb") as f:
        pickle.dump(imu_scaler, f)
        
    # PyTorch Tensor yapısına dönüştürme
    X_imu_tensor = torch.tensor(scaled_imu, dtype=torch.float32)
    
    # DataLoader hazırlığı (Mini-batch eğitim için)
    imu_dataset = TensorDataset(X_imu_tensor, X_imu_tensor) # Autoencoder girdi=çıktı mantığı
    imu_loader = DataLoader(imu_dataset, batch_size=256, shuffle=True)
    
    print("\n==========================================================================================")
    print("🏆 IMU VERİ KATMANI HAZIR!")
    print(f"✅ Tensor Boyutu: {X_imu_tensor.shape}")
    print(f"📍 Scaler diske mühürlendi: models/imu_scaler.pkl")
    print("==========================================================================================")
else:
    print("❌ Kritik Hata: İşlenebilecek uygun sensor_combined logu bulunamadı.")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import time

# ==========================================================================================
# 1. MODÜLER IMU AUTOENCODER MİMARİSİ
# ==========================================================================================
class IMUAutoencoder(nn.Module):
    def __init__(self):
        super(IMUAutoencoder, self).__init__()
        # Encoder: 6 boyuttan -> 3 boyuta sıkıştırma
        self.encoder = nn.Sequential(
            nn.Linear(6, 12),
            nn.ReLU(),
            nn.Linear(12, 3) # Gizli Temsil (Latent Space)
        )
        # Decoder: 3 boyuttan -> Tekrar 6 boyuta geri üretme
        self.decoder = nn.Sequential(
            nn.Linear(3, 12),
            nn.ReLU(),
            nn.Linear(12, 6),
            nn.Sigmoid() # Normalize veriyi [0,1] arasında tutmak için
        )

    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x

# Modeli oluştur ve eğer ekran kartın varsa GPU'ya (CUDA) taşı, yoksa CPU'da kalsın
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = IMUAutoencoder().to(device)

# Hata Fonksiyonu ve Optimize Edici
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

print(f"🧠 IMU Autoencoder Modeli {device} üzerinde oluşturuldu.")
print("🏋️ İleri Eğitim (Fine-Tuning / Klon Eğitimi) Başlıyor...\n")

# ==========================================================================================
# 2. MODEL EĞİTİM DÖNGÜSÜ (EPOCH = 5) - Veri çok büyük olduğu için 5 epoch yeterli kanka
# ==========================================================================================
num_epochs = 5

for epoch in range(num_epochs):
    start_time = time.time()
    train_loss = 0.0
    
    # imu_loader'ı bir önceki hücreden otomatik miras alıyor kanka
    for data, _ in imu_loader:
        data = data.to(device)
        
        # Gradyanları sıfırla
        optimizer.zero_grad()
        
        # İleri besleme (Forward)
        outputs = model(data)
        loss = criterion(outputs, data)
        
        # Geri besleme (Backward + Optimize)
        loss.backward()
        optimizer.step()
        
        train_loss += loss.item() * data.size(0)
        
    epoch_loss = train_loss / len(imu_loader.dataset)
    elapsed_time = time.time() - start_time
    
    print(f"Epoch [{epoch+1}/{num_epochs}] -> Ortalama Hata (Loss): {epoch_loss:.6f} | Süre: {elapsed_time:.2f} saniye")

# ==========================================================================================
# 3. KLON MODELİ KAYDETME
# ==========================================================================================
klon_model_yolu = r"C:\Users\feyza\Desktop\uav_project\models\imu_autoencoder_v2_tuning.pth"
torch.save(model.state_state(), klon_model_yolu) if hasattr(model, 'state_dict') else torch.save(model.state_dict(), klon_model_yolu)

print("\n==========================================================================================")
print("🏆 1. MODÜL (IMU) KLON EĞİTİMİ BAŞARIYLA TAMAMLANDI!")
print(f"📍 Yeni Kararlı Ağırlık Matrisi Kaydedildi: models/imu_autoencoder_v2_tuning.pth")
print("==========================================================================================")
```
